What is a probability of me randomly picking a card from 1 to 600 3 times and each time the card number is bigger than the previous one?

```python
# Simulate the problem withouth replacement.
cards = np.arange(0,600)
j=0
N=100000
for i in range (N):
    cards2 = cards.copy()
    x1=random.choice(cards2)
    cards2 = np.delete(cards2, np.where(cards2 == x1))
    x2=random.choice(cards2)
    cards2 = np.delete(cards2, np.where(cards2 == x2))
    x3=random.choice(cards2)
    cards2 = np.delete(cards2, np.where(cards2 == x3))
    if x3>x2>x1:
        j+=1

print(j/N)
```

```python
# Simulate the problem with replacement
j=0
N=100000
for i in range (N):
    cards2 = cards.copy()
    x1=random.choice(cards2)
    x2=random.choice(cards2)
    x3=random.choice(cards2)
    if x3>x2>x1:
        j+=1

print(j/N)
```

```python
from scipy.stats import uniform

hht_count = 0
htt_count = 0
N = 100000

for i in range(N):
    st = ''
    while True:
        flip = uniform.rvs()
        ch='H' if flip<0.5 else 'T'
        st+=ch
        if st[-3:] == 'HHT':
            hht_count+=1
            break
        elif st[-3:] =='HTT':
            htt_count+=1
            break
    
        
    
print(hht_count/N)
print(htt_count/N)
```
